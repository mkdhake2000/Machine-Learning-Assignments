Calculating Euclidean Distance

```python
def euclidis(x,y):
    dis=0
    for i in range(len(x)-1):
        dis+=((x[i]-y[i])**2);
    return dis**(1/2)
```

KNN

```python
import operator
def getNeighbors(trainset,testinst,k):
    dist=[]
    
    for x in trainset.index:
        dis=euclidis(trainset.loc[x],testinst)
        dist.append((x,dis))
    dist.sort(key=operator.itemgetter(1))
    neighbors=[]
    for x in range(k):
        i=dist[x][0]
        ans=trainset.loc[i,"price"]
        neighbors.append(ans)
    return neighbors
```

Returning the average of KNN

```python
def knnfunct(trainset,testset,k):
    result=[]
    for x in testset.index:
        neighbors=getNeighbors(trainset,testset.loc[x],k)
        avg=npy.mean(neighbors)
        result.append(avg)
    return result
```

Root Mean Square Error(RMSE) Calculation

```python
def rmse(l1,l2):
    l=len(l1)
    sm=0
    for x in l1.index:
        sm+=((l1[x]-l2[x])**2)
    sm=sm/l
    return sm**(1/2)
```

Defining Function to get RMSE

```python
def knnmodel(train_set,test_set,k):
    ans={}
    l1=test_set["price"]
    l1=l1.reset_index(drop=True)
    for x in range(1,k+1):
        l2=knnfunct(train_set,test_set,x)
        ans.update({"k"+str(x):rmse(l1,l2)})
        
    return ans
```

Using Cross-Validation to figure out best value of k

```python
result=knnmodel(train_set,cval_set,5)
result
```

Best Value for k is 1

So using KNN on test data with k=1

```python
res=knnmodel(train_set,test_set,1)
res
```

Defining a Node class denoting nodes in the tree

```python
class Node():
    def __init__(self, feat_in=None, thresh=None, left=None, right=None, i_g=None, val=None):
        #constructor 
        
        # for decision node
        self.feat_in = feat_in
        self.thresh = thresh
        self.left = left
        self.right = right
        self.i_g = i_g
        
        # for leaf node
        self.val = val
```

Creating a DT Classifer Class to do the classification(Most important methods are included here)

```python
class DTClassifier():
    def __init__(self, min_samples=2, depth=2):
        #constructor
        
        # initialize the root of the tree 
        self.root = None
        
        # stopping conditions
        self.min_samples = min_samples
        self.depth = depth
        
    def build_tree(self, ds, depth_n=0):
        #recursive function to build the tree
        
        X, Y = ds[:,:-1], ds[:,-1]
        nsamples, nfeatures = npy.shape(X)
        
        # split until stopping conditions are met
        if nsamples>=self.min_samples and depth_n<=self.depth:
            # find the best split
            best_split = self.get_split(ds, nsamples, nfeatures)
            # check if information gain is positive
            if best_split["i_g"]>0:
                # recur left
                lsubtree = self.build_tree(best_split["lds"], depth_n+1)
                # recur right
                rsubtree = self.build_tree(best_split["rds"], depth_n+1)
                # return decision node
                return Node(best_split["feat_in"], best_split["thresh"], 
                            lsubtree, rsubtree, best_split["i_g"])
        
        # compute leaf node
        leaf_val = self.calculate_leaf(Y)
        # return leaf node
        return Node(val=leaf_val)
    
    def get_split(self, ds, nsamples, nfeatures):
        #function to find the best split
        
        # dictionary to store the best split
        best_split = {}
        max_ig = -float("inf")
        
        # loop over all the features
        for feat_in in range(nfeatures):
            feat_val = ds[:, feat_in]
            possible_thresh = npy.unique(feat_val)
            # loop over all the feature values present in the data
            for thresh in possible_thresh:
                # get current split
                lds, rds = self.split(ds, feat_in, thresh)
                # check if childs are not null
                if len(lds)>0 and len(rds)>0:
                    y, ly, ry = ds[:, -1], lds[:, -1], rds[:, -1]
                    # compute information gain
                    curr_ig = self.i_g(y, ly, ry)
                    # update the best split if needed
                    if curr_ig>max_ig:
                        best_split["feat_in"] = feat_in
                        best_split["thresh"] = thresh
                        best_split["lds"] = lds
                        best_split["rds"] = rds
                        best_split["i_g"] = curr_ig
                        max_ig = curr_ig
                        
        # return best split
        return best_split
    
    def split(self, ds, feat_in, thresh):
        #function to split the data
        
        lds = npy.array([row for row in ds if row[feat_in]<=thresh])
        rds = npy.array([row for row in ds if row[feat_in]>thresh])
        return lds, rds
    
    def i_g(self, par, lchild, rchild):
        #function to compute information gain
        
        lweight = len(lchild) / len(par)
        rweight = len(rchild) / len(par)
        gain = self.entropy(par) - (lweight*self.entropy(lchild) + rweight*self.entropy(rchild))
        return gain
    
    def entropy(self, y):
        #function to compute entropy
        
        clabels = npy.unique(y)
        entropy = 0
        for c in clabels:
            pc = len(y[y == c]) / len(y)
            entropy += -pc * npy.log2(pc)
        return entropy
        
    def calculate_leaf(self, Y):
        #function to compute leaf node
        
        Y = list(Y)
        return max(Y, key=Y.count)
    
    def print_tree(self, tree=None, indent=" "):
        #function to print the tree
        
        if not tree:
            tree = self.root

        if tree.val is not None:
            print(tree.val)

        else:
            print("X_"+str(tree.feat_in), "<=", tree.thresh, "?", tree.i_g)
            print("%sleft:" % (indent), end="")
            self.print_tree(tree.left, indent + indent)
            print("%sright:" % (indent), end="")
            self.print_tree(tree.right, indent + indent)
    
    def fit(self, X, Y):
        #function to train the tree
        
        ds = npy.concatenate((X, Y), axis=1)
        self.root = self.build_tree(ds)
    
    def predict(self, X):
        #function to predict new dataset
        
        predictions = [self.predict_val(x, self.root) for x in X]
        return predictions
    
    def predict_val(self, x, tree):
        #function to predict a single data point 
        
        if tree.val!=None: return tree.val
        feature_val = x[tree.feat_in]
        if feature_val<=tree.thresh:
            return self.predict_val(x, tree.left)
        else:
            return self.predict_val(x, tree.right)
```
