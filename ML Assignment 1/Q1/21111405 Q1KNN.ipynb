Calculating Euclidean Distance

```python
def euclidis(x,y):
    dis=0
    for i in range(len(x)-1):
        dis+=((x[i]-y[i])**2);
    return dis**(1/2)
```

KNN

```python
import operator
def getNeighbors(trainset,testinst,k):
    dist=[]
    
    for x in trainset.index:
        dis=euclidis(trainset.loc[x],testinst)
        dist.append((x,dis))
    dist.sort(key=operator.itemgetter(1))
    neighbors=[]
    for x in range(k):
        i=dist[x][0]
        ans=trainset.loc[i,"price"]
        neighbors.append(ans)
    return neighbors
```

Returning the average of KNN

```python
def knnfunct(trainset,testset,k):
    result=[]
    for x in testset.index:
        neighbors=getNeighbors(trainset,testset.loc[x],k)
        avg=npy.mean(neighbors)
        result.append(avg)
    return result
```

Root Mean Square Error(RMSE) Calculation

```python
def rmse(l1,l2):
    l=len(l1)
    sm=0
    for x in l1.index:
        sm+=((l1[x]-l2[x])**2)
    sm=sm/l
    return sm**(1/2)
```

Defining Function to get RMSE

```python
def knnmodel(train_set,test_set,k):
    ans={}
    l1=test_set["price"]
    l1=l1.reset_index(drop=True)
    for x in range(1,k+1):
        l2=knnfunct(train_set,test_set,x)
        ans.update({"k"+str(x):rmse(l1,l2)})
        
    return ans
```

Using Cross-Validation to figure out best value of k

```python
result=knnmodel(train_set,cval_set,5)
result
```

Best Value for k is 1

So using KNN on test data with k=1

```python
res=knnmodel(train_set,test_set,1)
res
```
